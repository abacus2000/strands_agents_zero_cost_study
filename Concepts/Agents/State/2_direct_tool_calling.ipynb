```python
from strands import Agent
from strands.models.ollama import OllamaModel
from strands_tools import calculator
import pprint 
```

```python
model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)

agent = Agent(
    model=model,
    system_prompt="""You are a helpful assistant who is also a pompous mathmatical wizard.""",
    tools=[calculator]
)
```

# Direct Tool Call

```python
agent.tool.calculator(expression="123 * 456")
```

```python
pprint.pprint(agent.messages)
```

```python
# Direct tool call without recording
agent.tool.calculator(expression="765 / 987", record_direct_tool_call=False)
```

```python
pprint.pprint(agent.messages)
```

```python
from strands import Agent
from strands.models.ollama import OllamaModel
import pprint # for printing more readable outputs
```

# Basic Message

```python
# # Create an agent
# agent = Agent()

model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)

basic_agent = Agent(
    model=model,
    system_prompt="""You are a helpful assistant who talks like a pirate."""
)
```

```python
pprint.pprint(basic_agent("Hello!"))
```

```python
# Access the conversation history
pprint.pprint(basic_agent.messages)  # Shows all messages exchanged so far
```

# Initialize Agent With Existing Messages

```python

model_with_messages = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)

basic_agent_with_messages = Agent(
    model=model_with_messages,
    messages=[{"role": "user", "content": [{"text": "Hello, my name is Alex!"}]}],
    system_prompt="""You are a helpful assistant who talks like a detective from the Lovecraftian universe."""
)
```

```python
pprint.pprint(basic_agent_with_messages("what is my name"))
```

## Extracting Result

```python
result = basic_agent_with_messages("what is my name")
```

```python
result = basic_agent_with_messages("what is my name")
```

```python
pprint.pprint(result.to_dict())
```

```python
pprint.pprint(basic_agent_with_messages.messages) 
```

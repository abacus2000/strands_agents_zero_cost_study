```python
from strands import Agent
from strands.models.ollama import OllamaModel
from strands.agent.conversation_manager import SlidingWindowConversationManager
import pprint 
```

```python
model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)
```

```python

# Create a conversation manager with custom window size
# By default, SlidingWindowConversationManager is used even if not specified
conversation_manager = SlidingWindowConversationManager(
    window_size=4,  # Maximum number of message pairs to keep - includes both User and agent responses
)

# Use the conversation manager with your agent
agent_detective = Agent(
    model=model,
    system_prompt="""You are a helpful assistant who talks like a detective from the Lovecraftian universe.""",
    conversation_manager=conversation_manager
)

result = agent_detective("have I seen you before?")
```

```python
result = agent_detective("I have no memory of you, but you seem familiar. But I find I have a fleeting memory these days.")
```

```python
result = agent_detective("I have no memory of ... but (collapses to the floor)")
```

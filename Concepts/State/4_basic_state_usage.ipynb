```python
from strands import Agent
from strands.models.ollama import OllamaModel
import pprint 

model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)
```

```python
agent = Agent(model=model, state={"user_preferences": {"theme": "dark"}, "session_count": 0})


# Access state values
theme = agent.state.get("user_preferences")
print(theme)  # {"theme": "dark"}
```

```python
# Set new state values
agent.state.set("last_action", "login")
agent.state.set("session_count", 1)

# Get entire state
all_state = agent.state.get()
pprint.pprint(all_state)  # All state data as a dictionary
```

```python
# Delete state values
agent.state.delete("last_action")
all_state = agent.state.get()
```

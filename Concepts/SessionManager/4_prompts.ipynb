```python
from strands import Agent
from strands.models.ollama import OllamaModel


model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.2-vision:11b",
)

agent = Agent(
    model=model, 
    name="image_researcher"
)
from pathlib import Path

image_path = Path(
    "~/workplace/strands_agents_zero_cost_study/Concepts/SessionManager/images/test.jpg"
).expanduser()

with image_path.open("rb") as fp:
    image_bytes = fp.read()

    
response = agent([
    {"text": "What can you see in this image?"},
    {
        "image": {
            "format": "png",
            "source": {
                "bytes": image_bytes,
            },
        },
    },
])
```

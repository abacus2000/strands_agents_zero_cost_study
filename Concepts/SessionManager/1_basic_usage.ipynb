```python
from strands import Agent
from strands.session.file_session_manager import FileSessionManager
from strands.models.ollama import OllamaModel
import json

model = OllamaModel(
    host="http://localhost:11434",
    model_id="llama3.1",
)

# ==========================================
# CONVERSATION 1: Create new session
# ==========================================
print("=== Starting New Conversation ===\n")

# Create a session manager with unique session ID
session_manager = FileSessionManager(
    session_id="user-john-123",
    storage_dir="./agent_sessions"
)

# Create agent with session manager
agent = Agent(
    model=model,
    session_manager=session_manager,
    system_prompt="You are a helpful AI assistant. Remember user preferences."
)
```

```python
# First conversation
response1 = agent("Hi! My name is Alex and I love Python and Java programming.")
```

```python
response2 = agent("I am interested in learning how to contribute to open source, can you help with some tips?")
```

```python
# ==========================================
# CONVERSATION 2: Restore from session
# ==========================================
print("\n=== Restoring Previous Conversation ===\n")

# Simulate app restart - create new agent instance with SAME session_id
# This automatically restores all previous messages and state
restored_session_manager = FileSessionManager(
    session_id="user-john-123",  # Same session ID!
    storage_dir="./agent_sessions"
)

restored_agent = Agent(
    model=model,
    session_manager=restored_session_manager,
    system_prompt="You are a helpful AI assistant. Remember user preferences."
)

# Agent remembers previous conversation!
response3 = restored_agent("What's my name and what language do I like?")
```

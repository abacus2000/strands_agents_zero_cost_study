```python

from strands import Agent, tool
from strands.models import AnthropicModel
import os
```

```python
@tool
def add(a: float, b: float) -> float:
    """
    Add two numbers together.

    Args:
        a: The first number to add
        b: The second number to add

    Returns:
        The sum of a and b
    """
    # Simple addition operation
    result = a + b
    print(f"[Tool Executing] add({a}, {b}) = {result}")
return result


@tool
def subtract(a: float, b: float) -> float:
"""
Subtract the second number from the first number.

Args:
    a: The number to subtract from
    b: The number to subtract

Returns:
    The difference of a minus b
"""
# Simple subtraction operation
result = a - b
print(f"[Tool Executing] subtract({a}, {b}) = {result}")
return result


@tool
def multiply(a: float, b: float) -> float:
"""
Multiply two numbers together.

Args:
    a: The first number to multiply
    b: The second number to multiply

Returns:
    The product of a and b
"""
# Simple multiplication operation
result = a * b
print(f"[Tool Executing] multiply({a}, {b}) = {result}")
return result
```


#Using @tool Decorator

```python
model = AnthropicModel(model_name="claude-3-5-sonnet-20241022")

agent = Agent(
    model=model,
    tools=[add, subtract, multiply],  
    system_prompt="You are a helpful calculator assistant. Use the provided tools to perform calculations."
)
```

```python

prompt1 = "What is 25 plus 17?"

result1 = agent.run(prompt1)
```

```python

prompt2 = "Calculate (15 + 8) * 3"

result2 = agent.run(prompt2)
```

```python

prompt3 = "If I have 100 dollars and spend 37 dollars, how much do I have left?"

result3 = agent.run(prompt3)


```
